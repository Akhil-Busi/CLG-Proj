## рҹҺҜ Quick Start (5 minutes)

### Step 1: Install Dependencies
```bash
cd fusedchat_backend
pip install -r requirements.txt
```

### Step 2: Start Backend
```bash
cd fusedchat_backend
uvicorn app.main:app --reload --port 8000
```

### Step 3: Start Frontend (in another terminal)
```bash
cd fusedchat_frontend
npm start  # or PORT=3001 npm start
```

### Step 4: Open Browser
```
http://localhost:3001
```

````python
# Test 2: Professional Brain - Fast Mode
import requests
from uuid import uuid4

session_id = str(uuid4())
query = "What is a Linked List?"

payload = {
    "session_id": session_id,
    "query": query,
    "mode": "fast"
}

try:
    response = requests.post(f"{API_URL}/chat", json=payload)
    result = response.json()
    
    print("вң… Professional Brain (Fast Mode) Response:")
    print(f"Query: {query}")
    print(f"Brain: {result.get('brain')}")
    print(f"Answer: {result.get('response')[:200]}...")
    print(f"Confidence: {result.get('confidence')}")
except Exception as e:
    print(f"вқҢ Error: {e}")
````

## рҹ”§ Troubleshooting

### Problem: "ModuleNotFoundError: No module named 'app'"
**Solution:** Run from project root:
```bash
cd fusedchat_backend
python -m app.services.professional_brain
```

### Problem: "FAISS Index Not Found"
**Solution:** The syllabus index will be built automatically on first run if syllabus PDF is in `data/syllabus/`

### Problem: "Groq API Error"
**Solution:** Set your Groq API key in `.env`:
```
GROQ_API_KEY=your_key_here
```

### Problem: MongoDB Connection Error
**Solution:** Either:
- Install MongoDB locally
- Or use MongoDB Atlas (cloud) with connection string in `.env`

### Problem: React shows blank page
**Solution:** Check browser console (F12) for errors and ensure backend is running on port 8000
